```python
img	=cv2.imread('image.jpeg',cv2.IMREAD_GRAYSCALE)
hist=cv2.calcHist([img],[0],None,[128],[0,128])
equalized_img=cv2.equalizeHist(img)
equalized_hist=	cv2.calcHist([equalized_img],[0],None,[128],[0,128])
fig,axs	=plt.subplots(nrows=2,	ncols=2,	figsize=(10,10))
axs[0][0].imshow(img,	cmap='gray')
axs[0][0].set_title('Gray	Image')
axs[0][1].plot(hist)
axs[0][1].set_title('Histogram')
axs[1][0].imshow(equalized_img,	cmap='gray')
axs[1][0].set_title('Equalized	Image')
axs[1][1].plot(equalized_hist)
axs[1][1].set_title('Equalized	Histogram')
plt.tight_layout()
plt.show()
```
